# Regression Promblem

## load sample dataset

```python
from pycaret.datasets import get_data
data = get_data('insurance')
```

## Functional API（函式式寫法）

```python
from pycaret.regression import *
s = setup(data, target = 'charges', session_id = 123)
```

### 📊 資料摘要
| 描述                        | 說明                           |
|-----------------------------|--------------------------------|
| Session id                 | 123（隨機種子，讓結果可重現） |
| Target                     | `charges`（要預測的欄位）      |
| Target type                | Regression（回歸任務）         |
| Original data shape        | (1338, 7) 原始資料共 7 欄       |
| Transformed data shape     | (1338, 10) 經 one-hot 編碼擴展為 10 欄 |
| Transformed train set      | (936, 10) 訓練集                |
| Transformed test set       | (402, 10) 測試集                |

---

### 🧮 特徵工程與預處理
| 描述                        | 說明                               |
|-----------------------------|------------------------------------|
| Numeric features           | 3 個數值欄位                        |
| Categorical features       | 3 個類別欄位                        |
| Preprocess                 | True，已進行預處理                  |
| Imputation type            | simple（簡單補值）                  |
| Numeric imputation         | mean（數值以平均數補值）           |
| Categorical imputation     | mode（類別以眾數補值）             |
| Maximum one-hot encoding   | 最大 one-hot 類別數為 25，超過會跳過 |
| Encoding method            | 預設自動選擇（None）               |

---

### 🧪 交叉驗證與資源
| 描述                        | 說明                     |
|-----------------------------|--------------------------|
| Fold Generator             | 使用 KFold 交叉驗證     |
| Fold Number                | 分成 10 折              |
| CPU Jobs                   | -1（使用所有 CPU 核心） |
| Use GPU                    | False，未使用 GPU        |

---

### 📦 其他設定
| 描述                        | 說明                          |
|-----------------------------|-------------------------------|
| Log Experiment             | False（未記錄至 MLflow）     |
| Experiment Name            | `reg-default-name`            |
| USI                        | `ace5`（唯一實驗識別碼）      |

## Compare Models

```python
# functional API
best = compare_models()
print(best)
```

最佳模型：Gradient Boosting Regressor (gbr)
根據你畫黃的欄位：

MAE = 2701.99（最低）

MSE = 23548981.36（最低）

RMSE = 4832.97（最低）

R² = 0.8320（最高）

RMSLE = 0.4447（最低之一）

MAPE = 0.3137（也是最低）

✅ 綜合指標來看，gbr 是最佳模型，泛化能力強、誤差小。


## Analyze Model

```python
# functional API
evaluate_model(best)
```

| 視覺化選項                   | 說明                               |
| ----------------------- | -------------------------------- |
| `Pipeline Plot`         | 顯示整個前處理 + 模型的 pipeline 結構        |
| `Hyperparameters`       | 顯示目前模型的參數設定                      |
| `Residuals`             | 預測值與實際值的殘差圖                      |
| `Prediction Error`      | 預測值 vs 實際值的散佈圖（應接近 y=x）          |
| `Cooks Distance`        | 用來偵測影響力大的離群值（influential points） |
| `Feature Selection`     | 模型選用的特徵資訊                        |
| `Learning Curve`        | 訓練資料增加時的得分變化                     |
| `Manifold Learning`     | 非線性降維可視化（例如 t-SNE）               |
| `Validation Curve`      | 單一超參數值 vs 模型表現                   |
| `Feature Importance`    | 特徵重要性條形圖（某些模型支援）                 |
| `Decision Tree`         | 如果模型是樹類模型，顯示決策樹結構                |
| `Interactive Residuals` | 可互動的殘差分析圖（Hover 看數值）             |
