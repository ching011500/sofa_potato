Adding / Removing Columns

```python
# adding
coffee_new = coffee.copy()
coffee_new['Price'] = 4.99
coffee_new.head()
```

```python
import numpy as np 
coffee_new['New Price'] = np.where(coffee_new['Coffee Type'] == 'Espresso', 3.99, 5.99)
coffee_new
```

```python
# removing 
# coffee_new.drop(columns = ['Price'], inplace = True)
coffee_new
```

```python
coffee_new['Revenue'] = coffee_new['Units Sold'] * coffee_new['New Price']
coffee_new.head()
```

```python
# Rename
coffee_new.rename(columns = {'New Price' : 'Price'})
```

```python
bios_new = bios.copy()
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
bios_new.query('first_name == "Keith"')
```

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'])
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new
```

```python
# .apply() / lambda 
bios['height category'] = bios['height_cm'].apply(lambda x : 'Short' if x < 165 else('Average' if x < 185 else 'Tall'))
bios.head()
```

```python
# def
def categorize_athlete(row) : 
    if row['height_cm'] < 175 and row['weight_kg'] < 70 :
        return 'Lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] <= 80 :
        return 'Middleweight'
    else :
        return 'Heavyweight'
bios['Category'] = bios.apply(categorize_athlete, axis = 1)

bios.head()
```

Aggregating

```python
# .value_counts()
bios['born_city'].value_counts()
```

```python
bios[bios['born_country'] == 'USA']['born_region'].value_counts()
```

```python
# .groupby()
coffee_new.groupby('Coffee Type')['Units Sold'].sum()
```

```python
coffee_new.groupby(['Coffee Type', 'Day']).agg({'Units Sold' : 'sum', 'Price' : 'mean'})
```

```python
# pivot table
pivot = coffee_new.pivot(columns = 'Coffee Type', index = 'Day', values = 'Revenue')
pivot
```

```python
pivot.sum(axis = 1)
```

```python
# groupby combined with datetime operations
bios['born_date'] = pd.to_datetime(bios['born_date'])
bios.groupby(bios['born_date'].dt.year)['name'].count().reset_index().sort_values('name', ascending = False).head()
```

Advanced funcitonality

```python
# .shift()
coffee_new['yesterday_revenue'] = coffee_new['Revenue'].shift(2)
coffee_new.head()
```

```python
coffee_new['pct_change'] = coffee_new['Revenue'] / coffee_new['yesterday_revenue'] * 100
coffee_new.head()
```

```python
# .rank()
bios['Height_rank'] = bios['height_cm'].rank(ascending = False)
bios.sort_values('Height_rank', ascending=True).head()
```

```python
bios.sort_values(['Height_rank']).sample(10)[['name', 'Height_rank']]
```

```python
# .cumsum()
coffee_new.select_dtypes('float').cumsum().head()
```

```python
coffee_new['cumulative_revenue'] = coffee_new['Revenue'].cumsum()
coffee_new.head()
```

```python
# rolling
latte = coffee_new[coffee_new['Coffee Type'] == 'Latte'].copy()
latte['3Day'] = latte['Units Sold'].rolling(3).sum()
latte[['Units Sold', '3Day']]
```
